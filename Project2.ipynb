```python
#Zening Li
#import and read data
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
df = pd.read_csv("NCWIT_DataV2_RawData.csv")
df

#clear data
df.rename(index=str, columns={"Record#": "Record Number", "CIP# Only": "CIP Number Only"})
df.fillna(0) #fill NaN to 0

#visualization 1
df["Average ACT Score"] = (df["Enroll, Female: Avg. ACT Math Score (Enrl F)"] + df["Enroll, Male: Avg. ACT Math Score (Enrl M)"])/2
df["Average HS GPA"] = (df["Enroll, Female: Average HS GPA (Enrl F)"] + df["Enroll, Male: Average HS GPA (Enrl M)"])/2
my_bins = range(15,35,3)
fig, ax = plt.subplots(figsize=(8,4))
df.hist(column="Average ACT Score", ax=ax, bins=my_bins, stacked=True, facecolor="steelblue", edgecolor="white")
df.hist(column="Average HS GPA", ax=ax, bins=my_bins, stacked=True, facecolor="green", edgecolor="white")
ax.set_title("Average ACT Score", fontsize=20)
ax.set_xlabel("Score", fontsize=16)
ax.set_ylabel("Frequency", fontsize=16)






```
