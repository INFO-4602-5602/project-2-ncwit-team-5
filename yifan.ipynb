```python
import pandas as pd
data = pd.read_csv("NCWIT_DataV2_RawData.csv")
# data.dropna(axis=0)
data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
one = data['When do students typically declare their major?']
finish = one.dropna()
fig, ax = plt.subplots()
finish.value_counts().plot(ax=ax, kind='bar')
plt.title('Frequency for declearing major')
plt.show()
```
